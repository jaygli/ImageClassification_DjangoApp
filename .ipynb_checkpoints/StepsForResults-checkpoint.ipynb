```python
from keras.preprocessing import image
import tensorflow as tf
from tensorflow import Graph
```

```python
img_height, img_width=224,224
```

```python
with open('models/imagenet_classes.json', 'r') as f:
    labelInfo=f.read()
```

```python
import json
```

```python
labelInfo=json.loads(labelInfo)
```

```python
testimage='media/1.png'
```

```python
from keras.applications import mobilenet
```

```python
model=mobilenet.MobileNet(include_top=True,weights='imagenet')
```

```python
model.layers
```

```python
img = image.load_img(testimage, target_size=(img_height, img_width))
```

```python
x = image.img_to_array(img)
x=x/255
x=x.reshape(1, img_height, img_width, 3)
```

```python
predi=model.predict(x)
```

```python
predi
```

```python
for idx, x in np.ndenumerate(predi[0]):
    print (labelInfo[str(idx[0])][1], x)
```

```python
labelInfo['0']
```

```python
results = []
for idx, x in np.ndenumerate(predi[0]):
    results.append ({"label": labelInfo[str(idx[0])][1], "prediction": x})
```

```python
results
```

```python
sortedResults = sorted(results, key=lambda x: x["prediction"], reverse=True)
```
